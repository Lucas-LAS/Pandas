## Primeiro Comando

>**pd.read_csv** Ler o arquvio em formato CSV
**columns** Exibir os nomes da colunas e também pode ser usado para alterar os nomes das colunas
**head()** Definir quantas linhas serão exibidas, por padrão são exibidas 5 primeiras linhas.


```python
link_notas ='https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula4.1/ratings.csv'
notas = pd.read_csv(link_notas)
notas.columns = ['ID_Usuario','Filme','Nota','Data']
notas.head(10)

```

```python
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'red' if val < 3 else 'black'
    return 'color: %s' % color
```

```python
s = notas.head(100).style.applymap(color_negative_red)
s
```
